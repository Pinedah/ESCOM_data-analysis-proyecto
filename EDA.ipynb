## Gráficas

```python
def plot_numeric_distributions(df, save_individual=True):
    # Crear carpeta si no existe
    if save_individual and not os.path.exists('graficas'):
        os.makedirs('graficas')
    
    # Seleccionar solo columnas numéricas
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    # Calcular número de filas y columnas para subplots
    n_cols = 3
    n_rows = (len(numeric_cols) + n_cols - 1) // n_cols
    
    # Crear figura
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    # Graficar cada variable
    for idx, col in enumerate(numeric_cols):
        axes[idx].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'Distribución de {col}')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frecuencia')
        
        # Guardar gráfica individual
        if save_individual:
            fig_individual = plt.figure(figsize=(8, 6))
            plt.hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
            plt.title(f'Distribución de {col}')
            plt.xlabel(col)
            plt.ylabel('Frecuencia')
            plt.tight_layout()
            plt.savefig(f'graficas/{col}_distribucion.png', dpi=300, bbox_inches='tight')
            plt.close(fig_individual)
    
    # Ocultar subplots vacíos
    for idx in range(len(numeric_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Usar la función
plot_numeric_distributions(dataset_sinfn_clean)
```

```python
# Gráficas de distribución para variables categóricas (mejorado para muchas categorías)
def plot_categorical_distributions(df, save_individual=True, max_categories=10):
    # Crear carpeta si no existe
    if save_individual and not os.path.exists('graficas_cat'):
        os.makedirs('graficas_cat')
    
    # Seleccionar solo columnas categóricas
    categorical_cols = df.select_dtypes(include=['object']).columns
    
    # Calcular número de filas y columnas para subplots
    n_cols = 2
    n_rows = (len(categorical_cols) + n_cols - 1) // n_cols
    
    # Crear figura
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    # Graficar cada variable
    for idx, col in enumerate(categorical_cols):
        # Contar frecuencias
        value_counts = df[col].value_counts()
        
        # Limitar a las primeras max_categories categorías si hay muchas
        if len(value_counts) > max_categories:
            value_counts = value_counts.head(max_categories)
            title_suffix = f' (Top {max_categories})'
        else:
            title_suffix = ''
        
        # Crear gráfica de barras
        axes[idx].bar(range(len(value_counts)), value_counts.values, edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'Distribución de {col}{title_suffix}')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frecuencia')
        axes[idx].set_xticks(range(len(value_counts)))
        axes[idx].set_xticklabels(value_counts.index, rotation=45, ha='right', fontsize=8)
        
        # Guardar gráfica individual
        if save_individual:
            fig_individual = plt.figure(figsize=(10, 6))
            plt.bar(range(len(value_counts)), value_counts.values, edgecolor='black', alpha=0.7)
            plt.title(f'Distribución de {col}{title_suffix}')
            plt.xlabel(col)
            plt.ylabel('Frecuencia')
            plt.xticks(range(len(value_counts)), value_counts.index, rotation=45, ha='right', fontsize=9)
            plt.tight_layout()
            plt.savefig(f'graficas_cat/{col}_distribucion.png', dpi=300, bbox_inches='tight')
            plt.close(fig_individual)
    
    # Ocultar subplots vacíos
    for idx in range(len(categorical_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Usar la función

plot_categorical_distributions(dataset_sinfn_clean[['Estado', 'Region', 'Causa', 'Causa_especifica', 'Duracion_dias', 'Tipo_de_incendio', 'Tipo_Vegetacion', 'Regimen_de_fuego', 'Tipo_impacto', 'Tamano']])
```

## Outliers

```python
# Análisis de Outliers usando el método IQR (Rango Intercuartílico)
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar solo columnas numéricas
numeric_cols = dataset_sinfn_clean.select_dtypes(include=['int64', 'float64']).columns

print("=" * 80)
print("ANÁLISIS DE OUTLIERS POR COLUMNA (Método IQR)")
print("=" * 80)

# Diccionario para almacenar información de outliers
outliers_info = {}

for col in numeric_cols:
    # Calcular Q1, Q3 e IQR
    Q1 = dataset_sinfn_clean[col].quantile(0.25)
    Q3 = dataset_sinfn_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calcular límites
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identificar outliers
    outliers = dataset_sinfn_clean[(dataset_sinfn_clean[col] < lower_bound) | 
                                    (dataset_sinfn_clean[col] > upper_bound)]
    
    # Calcular porcentaje de outliers
    outlier_percentage = (len(outliers) / len(dataset_sinfn_clean)) * 100
    
    # Guardar información
    outliers_info[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'n_outliers': len(outliers),
        'percentage': outlier_percentage
    }
    
    # Imprimir información
    print(f"\n{col}:")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Límite inferior: {lower_bound:.2f}")
    print(f"  Límite superior: {upper_bound:.2f}")
    print(f"  Número de outliers: {len(outliers)} ({outlier_percentage:.2f}%)")

# Crear visualización: Boxplots para todas las columnas numéricas
n_cols = 3
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes

for idx, col in enumerate(numeric_cols):
    axes[idx].boxplot(dataset_sinfn_clean[col].dropna(), vert=True)
    axes[idx].set_title(f'Boxplot de {col}')
    axes[idx].set_ylabel(col)
    axes[idx].grid(True, alpha=0.3)
    
    # Añadir texto con información de outliers
    info = outliers_info[col]
    axes[idx].text(0.02, 0.98, 
                   f"Outliers: {info['n_outliers']} ({info['percentage']:.1f}%)",
                   transform=axes[idx].transAxes,
                   verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                   fontsize=8)
```

```python
def eliminar_outliers_por_umbral(df, umbrales, modo="ambos"):

    df_filtrado = df.copy()

    for col, umbral in umbrales.items():
        if col not in df_filtrado.columns:
            raise ValueError(f"La columna '{col}' no existe en el DataFrame")

        if modo == "superior":
            df_filtrado = df_filtrado[df_filtrado[col] <= umbral]

        elif modo == "inferior":
            df_filtrado = df_filtrado[df_filtrado[col] >= umbral]

        elif modo == "ambos":
            # asume umbral como valor absoluto máximo permitido
            df_filtrado = df_filtrado[df_filtrado[col].abs() <= umbral]

        else:
            raise ValueError("modo debe ser 'superior', 'inferior' o 'ambos'")

    return df_filtrado

umbrales = {
    "Deteccion": 1000000,       
    "Llegada": 1500000,
    "Duracion": 10000000    
}

dataset_sinfn_clean = eliminar_outliers_por_umbral(dataset_sinfn_clean, umbrales, modo="superior")
```

```python
# Función para normalizar con MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

def normalize_numeric_columns(df, columns=None):
    """
    Normaliza las columnas numéricas del dataframe usando MinMaxScaler
    
    Parameters:
    df: DataFrame con columnas numéricas y categóricas
    columns: lista de columnas a normalizar. Si es None, normaliza todas las columnas numéricas
    
    Returns:
    df_normalized: DataFrame con columnas numéricas normalizadas
    scaler: objeto MinMaxScaler ajustado (para poder revertir la transformación si es necesario)
    """
    # Crear copia del dataframe
    df_normalized = df.copy()
    
    # Seleccionar columnas a normalizar
    if columns is None:
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    else:
        numeric_cols = columns
    
    # Crear y ajustar el scaler
    scaler = MinMaxScaler()
    df_normalized[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    
    print(f"Columnas normalizadas: {list(numeric_cols)}")
    
    return df_normalized, scaler

# Aplicar normalización solo a columnas específicas
dataset_normalized, scaler = normalize_numeric_columns(
    dataset_sinfn_clean, 
    columns=['Arbolado_Adulto', 'Renuevo', 'Arbustivo', 'Herbaceo']
)
dataset_normalized.describe()
```

```python
# Análisis de Componentes Principales (PCA) con variables específicas
from sklearn.decomposition import PCA

# Seleccionar las variables específicas normalizadas para PCA
pca_columns = ['Arbolado_Adulto', 'Renuevo', 'Arbustivo', 'Herbaceo', 'Hojarasca', 'Total_hectareas']
numeric_data = dataset_normalized[pca_columns]

# Crear el modelo PCA con solo 2 componentes
pca = PCA(n_components=2)
pca_result = pca.fit_transform(numeric_data)

# Crear DataFrame con los componentes principales
pca_df = pd.DataFrame(
    data=pca_result,
    columns=['PC1', 'PC2']
)

# Agregar las columnas de Tipo_impacto y Duracion_dias para agrupar
pca_df['Tipo_impacto'] = dataset_sinfn_clean['Tipo_impacto'].values
pca_df['Duracion_dias'] = dataset_sinfn_clean['Duracion_dias'].values

# Convertir Duracion_dias a valores numéricos para ordenamiento
duracion_mapping = {
    '1 día': 1,
    '2 a 3 días': 2.5,
    '4 a 7 días': 5.5,
    'Mayor de 7 días': 10
}
pca_df['Duracion_dias_num'] = pca_df['Duracion_dias'].map(duracion_mapping)

# Varianza explicada por cada componente
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

print("Varianza explicada por cada componente:")
for i, var in enumerate(explained_variance):
    print(f"PC{i+1}: {var:.4f} ({var*100:.2f}%)")

print(f"\nVarianza acumulada total: {cumulative_variance[-1]:.4f} ({cumulative_variance[-1]*100:.2f}%)")

# Gráficas
fig = plt.figure(figsize=(20, 10))

# Gráfica de barras de varianza explicada
ax1 = plt.subplot(2, 2, 1)
ax1.bar(range(1, len(explained_variance)+1), explained_variance, alpha=0.7, edgecolor='black')
ax1.set_xlabel('Componente Principal')
ax1.set_ylabel('Varianza Explicada')
ax1.set_title('Varianza Explicada por Componente')

# Gráfica de varianza acumulada
ax2 = plt.subplot(2, 2, 2)
ax2.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o', linestyle='-', linewidth=2)
ax2.axhline(y=0.95, color='r', linestyle='--', label='95% varianza')
ax2.set_xlabel('Número de Componentes')
ax2.set_ylabel('Varianza Acumulada')
ax2.set_title('Varianza Acumulada')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Scatter plot PC1 vs PC2 agrupado por Tipo_impacto
ax3 = plt.subplot(2, 2, 3)
tipos_impacto = pca_df['Tipo_impacto'].unique()
colors_impacto = plt.cm.tab10(np.linspace(0, 1, len(tipos_impacto)))

for tipo, color in zip(tipos_impacto, colors_impacto):
    mask = pca_df['Tipo_impacto'] == tipo
    ax3.scatter(pca_df.loc[mask, 'PC1'], pca_df.loc[mask, 'PC2'], 
                label=tipo, alpha=0.6, s=30, color=color)

ax3.set_xlabel(f'PC1 ({explained_variance[0]*100:.2f}%)')
ax3.set_ylabel(f'PC2 ({explained_variance[1]*100:.2f}%)')
ax3.set_title('PCA: PC1 vs PC2 por Tipo de Impacto')
ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
ax3.grid(True, alpha=0.3)

# Scatter plot PC1 vs PC2 agrupado por Duracion_dias
ax4 = plt.subplot(2, 2, 4)
duraciones = pca_df['Duracion_dias'].unique()
colors_duracion = plt.cm.viridis(np.linspace(0, 1, len(duraciones)))

for duracion, color in zip(sorted(duraciones, key=lambda x: duracion_mapping.get(x, 0)), colors_duracion):
    mask = pca_df['Duracion_dias'] == duracion
    ax4.scatter(pca_df.loc[mask, 'PC1'], pca_df.loc[mask, 'PC2'], 
                label=duracion, alpha=0.6, s=30, color=color)

ax4.set_xlabel(f'PC1 ({explained_variance[0]*100:.2f}%)')
ax4.set_ylabel(f'PC2 ({explained_variance[1]*100:.2f}%)')
ax4.set_title('PCA: PC1 vs PC2 por Duración en Días')
ax4.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Mostrar los primeros componentes principales
print("\nPrimeros 5 registros de componentes principales:")
print(pca_df.head())

# Mostrar las cargas (loadings) de las variables originales en los componentes principales
loadings = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=pca_columns
)
print("\nCargas de las variables en los 2 componentes:")
print(loadings.round(3))
```
