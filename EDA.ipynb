Se procede a analizar las demás columnas con algunos valores faltantes, al representar estos un porcentaje tan bajo a la totalidad de los datos, se eliminarán dichos registros tal cual.

```python
# Eliminar registros con valores faltantes
dataset_sinfn_clean = dataset_sinfn.dropna()
print(f"Cantidad de valores faltantes:{dataset_sinfn.shape[0] - dataset_sinfn_clean.shape[0]} que represetan un {((dataset_sinfn.shape[0] - dataset_sinfn_clean.shape[0])/dataset_sinfn.shape[0])*100}% de los datos")
print(f"Forma antes de eliminar valores faltantes: {dataset_sinfn.shape}")
print(f"Forma después de eliminar valores faltantes: {dataset_sinfn_clean.shape}")
```

```python
for column in dataset_sinfn_clean.columns:
    print(f"{column} - {dataset_sinfn_clean[column].unique()}" if dataset_sinfn_clean[column].dtype == 'object' else column)
```

Analizamos si existe alguna otra forma de etiqutar algún valor perdido para las variables categoricas, la cuál sí existe, pero es representativa ya que explica desconocimiento no en el dato en sí, sino que la causa es desconocida.

```python
for column in dataset_sinfn_clean.columns:
    if dataset_sinfn_clean[column].dtype == 'object':
        dataset_sinfn_clean[column] = dataset_sinfn_clean[column].str.capitalize()
```

Se realiza una homologación en como están escritos los registros de las variables categoricas ya que eso daba diferentes valores unicos, y para las gráficas y análisis crearía categorias distintas.

```python
for column in dataset_sinfn_clean.columns:
    print(f"{column} - {dataset_sinfn_clean[column].unique()}" if dataset_sinfn_clean[column].dtype == 'object' else column)
```

```python
for column in dataset_sinfn_clean.columns:
    count_zeros = (dataset_sinfn_clean[column] == '0').sum()
    print(f"\t\t\t\t{column}: {count_zeros} ceros" if dataset[column].dtype == 'object' else f"{column} es numerica")
```

Se encuentra que existen registros en las variables categoricas etiquetados como "0", analizando la fuente de los datos y a que se refiere la variable en particular, se deduce que esta categoria refiere al desconocimiento del tipo de vegetación o la causa, por lo que se procederá a re etiquetar con "Desconocido"

```python
# Re-etiquetar valores '0' como 'Desconocido' en variables categóricas
for column in dataset_sinfn_clean.columns:
    if dataset_sinfn_clean[column].dtype == 'object':
        dataset_sinfn_clean[column] = dataset_sinfn_clean[column].replace('0', 'Desconocido')
```

## Gráficas

```python
def plot_numeric_distributions(df, save_individual=True):
    # Crear carpeta si no existe
    if save_individual and not os.path.exists('graficas'):
        os.makedirs('graficas')
    
    # Seleccionar solo columnas numéricas
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    # Calcular número de filas y columnas para subplots
    n_cols = 3
    n_rows = (len(numeric_cols) + n_cols - 1) // n_cols
    
    # Crear figura
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    # Graficar cada variable
    for idx, col in enumerate(numeric_cols):
        axes[idx].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'Distribución de {col}')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frecuencia')
        
        # Guardar gráfica individual
        if save_individual:
            fig_individual = plt.figure(figsize=(8, 6))
            plt.hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
            plt.title(f'Distribución de {col}')
            plt.xlabel(col)
            plt.ylabel('Frecuencia')
            plt.tight_layout()
            plt.savefig(f'graficas/{col}_distribucion.png', dpi=300, bbox_inches='tight')
            plt.close(fig_individual)
    
    # Ocultar subplots vacíos
    for idx in range(len(numeric_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Usar la función
plot_numeric_distributions(dataset_sinfn_clean)
```

```python
# Gráficas de distribución para variables categóricas
def plot_categorical_distributions(df, save_individual=True):
    # Crear carpeta si no existe
    if save_individual and not os.path.exists('graficas_categoricas'):
        os.makedirs('graficas_categoricas')
    
    # Seleccionar solo columnas categóricas
    categorical_cols = df.select_dtypes(include=['object']).columns
    
    # Calcular número de filas y columnas para subplots
    n_cols = 2
    n_rows = (len(categorical_cols) + n_cols - 1) // n_cols
    
    # Crear figura
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes] if n_cols == 1 else axes
    
    # Graficar cada variable
    for idx, col in enumerate(categorical_cols):
        # Contar frecuencias
        value_counts = df[col].value_counts()
        
        # Crear gráfica de barras
        axes[idx].bar(range(len(value_counts)), value_counts.values, edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'Distribución de {col}')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frecuencia')
        axes[idx].set_xticks(range(len(value_counts)))
        axes[idx].set_xticklabels(value_counts.index, rotation=45, ha='right')
        
        # Guardar gráfica individual
        if save_individual:
            fig_individual = plt.figure(figsize=(10, 6))
            plt.bar(range(len(value_counts)), value_counts.values, edgecolor='black', alpha=0.7)
            plt.title(f'Distribución de {col}')
            plt.xlabel(col)
            plt.ylabel('Frecuencia')
            plt.xticks(range(len(value_counts)), value_counts.index, rotation=45, ha='right')
            plt.tight_layout()
            plt.savefig(f'graficas/{col}_distribucion.png', dpi=300, bbox_inches='tight')
            plt.close(fig_individual)
    
    # Ocultar subplots vacíos
    for idx in range(len(categorical_cols), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Usar la función
plot_categorical_distributions(dataset_sinfn_clean)
```
